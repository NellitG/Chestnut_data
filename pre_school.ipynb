```python
from sqlalchemy import create_engine
import pandas as pd

# Create a connection to the MySQL database
engine = create_engine("mysql+mysqlconnector://root:@localhost/cape chestnut")

# Load a table into a pandas DataFrame
df = pd.read_sql("SELECT * FROM pre_school", engine)

df.head()
```

```python
#Line graph to show the trend of TERM_1_BALANCE
import matplotlib.pyplot as plt
if 'TERM_1_BALANCE' in df.columns:
    plt.plot(df['TERM_1_BALANCE'], marker ='o', linestyle='-', color='blue')
    plt.title('Trend of TERM_1_BALANCE')
    plt.xlabel('Index')
    plt.xticks(rotation=45)
    plt.ylabel('TERM_1_BALANCE')
    plt.grid(True)
    plt.show()
else:   
    print("Column 'TERM_1_BALANCE' does not exist.")
```

```python
#To generate PUPILS_NAME with TERM_1_BALANCE > 10000
df_filter = df[df['TERM_1_BALANCE'] > 10000]
df_filter
```

```python
#Generate a graph showing the trend of TERM_1_BALANCE > 10000
import matplotlib.pyplot as plt
if 'TERM_1_BALANCE' in df.columns:
    plt.plot(df_filter['TERM_1_BALANCE'], marker ='o', linestyle='-', color='red')
    plt.title('Trend of TERM_1_BALANCE > 10000')
    plt.xlabel('Index')
    plt.xticks(rotation=45)
    plt.ylabel('TERM_1_BALANCE')
    plt.grid(True)
    plt.show()
else:   
    print("Column 'TERM_1_BALANCE' does not exist.")
```
