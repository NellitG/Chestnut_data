```python
from sqlalchemy import create_engine
import pandas as pd

# Create a connection to the MySQL database
engine = create_engine("mysql+mysqlconnector://root:@localhost/cape chestnut")

# Load a table into a pandas DataFrame
df = pd.read_sql("SELECT * FROM upper", engine)

df.head()
```

```python
#Line graph to show the trend of Column_2
import matplotlib.pyplot as plt
if 'Column_2' in df.columns:
    plt.plot(df['Column_2'], marker ='o', linestyle='-', color='black')
    plt.title('Trend of Column_2')
    plt.xlabel('Index')
    plt.xticks(rotation=45)
    plt.ylabel('Column_2')
    plt.grid(True)
    plt.show()
else:   
    print("Column 'Column_2' does not exist.")
```

```python
#To generate upper with Column_2 > 10000
df_filter = df[df['Column_2'] > 10000]
df_filter
```

```python
#To generate graph of upper with Column_2 > 10000
import matplotlib.pyplot as plt
if 'Column_2' in df.columns:
    plt.plot(df_filter['Column_2'], marker ='o', linestyle='-', color='red')
    plt.title('Trend of Column_2')
    plt.xlabel('Index')
    plt.xticks(rotation=45)
    plt.ylabel('Column_2')
    plt.grid(True)
    plt.show()
else:   
    print("Column 'Column_2' does not exist.")
```
